```python
#concatenating along rows(axis=0)
import pandas as pd
df1=pd.DataFrame({
    'A':['a0','a1','a2'],
    'B':['b0','b1','b2']
})
print(df1)
df2=pd.DataFrame({
    'A':['a3','a4','a5'],
    'B':['b3','b4','b5']
})
print(df2)
res=pd.concat([df1,df2],ignore_index=True)
print(res)
```

```python
#concatenating along columns(axis=1)
import pandas as pd
df1=pd.DataFrame({
    'A':['a0','a1','a2'],
    'B':['b0','b1','b2']
})
print(df1)
df2=pd.DataFrame({
    'A':['a3','a4','a5'],
    'B':['b3','b4','b5']
})
print(df2)
res=pd.concat([df1,df2],axis=1)
print(res)
```

```python
#merge operations
#inner merge
import pandas as pd
df1=pd.DataFrame({'key':['a','b','c','d'],
                  'value':[1,2,3,4]})
df2=pd.DataFrame({'key':['b','d','e','f'],
                  'value':[5,6,7,8]})
res=pd.merge(df1,df2,on='key',how='inner')
print(res)
```

```python
#left merge
import pandas as pd
df1=pd.DataFrame({'key':['a','b','c','d'],
                  'value':[1,2,3,4]})
df2=pd.DataFrame({'key':['b','d','e','f'],
                  'value':[5,6,7,8]})
res=pd.merge(df1,df2,on='key',how='left')
print(res)
```

```python
#right merge
import pandas as pd
df1=pd.DataFrame({'key':['a','b','c','d'],
                  'value':[1,2,3,4]})
df2=pd.DataFrame({'key':['b','d','e','f'],
                  'value':[5,6,7,8]})
res=pd.merge(df1,df2,on='key',how='right')
print(res)
```

```python
#outer merge
import pandas as pd
df1=pd.DataFrame({'key':['a','b','c','d'],
                  'value':[1,2,3,4]})
df2=pd.DataFrame({'key':['b','d','e','f'],
                  'value':[5,6,7,8]})
res=pd.merge(df1,df2,on='key',how='outer')
print(res)
```
